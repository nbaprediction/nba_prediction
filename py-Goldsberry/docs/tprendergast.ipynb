One accessed the bits from the game they are interested in as follows: goldsberry.game.**class(GID)**.function()

```python
import pandas as pd
import goldsberry

pd.set_option("display.max_columns", 20)
```

```python
gid = '0021500550'
#gid is a string, not an integer
```

***Play_by_Play***

```python
df = pd.DataFrame(goldsberry.game.play_by_play(gid).plays())
df
```

***boxscore_summary***

```python
df = pd.DataFrame(goldsberry.game.boxscore_summary(gid).game_summary())
df     
```

```python
df = pd.DataFrame(goldsberry.game.boxscore_summary(gid).other_stats())
df     
```

```python
df = pd.DataFrame(goldsberry.game.boxscore_summary(gid).officials())
df
```

```python
df = pd.DataFrame(goldsberry.game.boxscore_summary(gid).game_info())
df
```

```python
df = pd.DataFrame(goldsberry.game.boxscore_summary(gid).last_meeting())
df
```

```python
df = pd.DataFrame(goldsberry.game.boxscore_summary(gid).season_series())
df
```

***boxscore_traditional***

```python
df = pd.DataFrame(goldsberry.game.boxscore_traditional(gid).player_stats())
df
```

```python
df = pd.DataFrame(goldsberry.game.boxscore_traditional(gid).team_stats())
df
```

***boxscore_advanced***

```python
df = pd.DataFrame(goldsberry.game.boxscore_advanced(gid).player_stats())
df
```

```python
df = pd.DataFrame(goldsberry.game.boxscore_advanced(gid).team_stats())
df
```
